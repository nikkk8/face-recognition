```python
import cv2
import os
import face_recognition as fr
import numpy as np
```

```python
path='images'
```

```python
mylist=os.listdir(path)
mylist
```

```python
imgs=[]
classnames=[]
for img in mylist:
    imgpath=os.path.join(path,img)
    print(imgpath)
    current_img=cv2.imread(imgpath)
    imgs.append(current_img)
    classnames.append(img.split(".")[0])
print(len(imgs))
print(classnames)
```

```python
def faceencoding(images):
    encodedlist=[]
    for img in images:
        img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame=fr.face_locations(img1)
        encode=fr.face_encodings(img1,face_in_frame)[0]
        #print(face_in_frame)
        #print(encode)
        encodedlist.append(encode)
    return encodedlist
```

```python
encoded_list_knownFaces=faceencoding(imgs)
encoded_list_knownFaces
```

```python
import winsound
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    face_in_frame=fr.face_locations(img1)
    # print(face_in_frame)
    encoded_faces=fr.face_encodings(img1,face_in_frame)
    # print(encoded_faces)
    for encfc,faceloc in zip(encoded_faces,face_in_frame):
        matches=fr.compare_faces(encoded_list_knownFaces,encfc)
        print(matches)
        facedistance=fr.face_distance(encoded_list_knownFaces,encfc)
        print(facedistance)
        matchind=np.argmin(facedistance)
        print(matchind)
        if matches[matchind]==True:
            name=classnames[matchind]
        else:
            name="Unknown"
            winsound.Beep(1000, 500)
            # print(name)
        y1,x2,y2,x1=faceloc
        cv2.rectangle(img,(x1, y1),(x2, y2),(0,255,0),2)
        cv2.putText(img,name,(x1, y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)


    cv2.imshow("WEBCAM",img)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
